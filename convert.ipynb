```python
import pandas as pd
import numpy as np
```

```python
beneficiario2021 = pd.read_csv('../Beneficiario2021.csv', sep=';')
```

```python
beneficiario2021.head()
```

```python
distance1 = pd.read_csv('../Distance1.csv')
```

```python
distance1.head()
```

```python
#merge datasets
mergeBenDis = pd.merge(beneficiario2021, distance1, on='WijkenEnBuurten', how='inner')
```

```python
mergeBenDis.head()
```

```python
mergeBenDis.columns
```

```python
#replace weird nan values for real nan values
mergeBenDis = mergeBenDis.replace('       .', np.nan)
```

```python
mergeBenDis.head()
```

```python
#get all columns that contain no values
columns_with_all_missing = mergeBenDis.columns[mergeBenDis.isnull().all()].tolist()
```

```python
print(columns_with_all_missing)
```

```python
#remove all columns that contain no values
newMergeBenDis = mergeBenDis.drop(columns=columns_with_all_missing)
```

```python
newMergeBenDis.head()
```

```python
#get duplicate column names
columns_x = [col for col in newMergeBenDis.columns if col.endswith('_x')]
columns_y = [col for col in newMergeBenDis.columns if col.endswith('_y')]
```

```python
print(columns_x)
print(columns_y)
```

```python
#check if the values in these columns are consistently the same, based on duplicate columns
for col_x, col_y in zip(columns_x, columns_y):
    
    are_equal = (newMergeBenDis[col_x] == newMergeBenDis[col_y]).all()

    if are_equal:
        print(f"The values in columns {col_x} and {col_y} are consistent.")
    else:
        print(f"The values in columns {col_x} and {col_y} are not consistent.")
```

### ID
ID_x and ID_y = logical -> datasets are from different cases with different ID values. Not important to consider.
### Gemeentenaam
Gemeentenaam_1_x and Gemeentenaam_1_y = weird; gemeentenaam should be the same in both cases. Let's validate

```python
#weird; gemeentenaam should be the same in both cases. Let's validate
gemeenteValueX = 'Gemeentenaam_1_x'
gemeenteValueY = 'Gemeentenaam_1_y'

for index, row in newMergeBenDis.iterrows():
    value_x = row[gemeenteValueX]
    value_y = row[gemeenteValueY]

    if value_x == value_y:
        print("looks good")
    else:
        print(f"look out for: {index} with {value_x} and {value_y} in {gemeenteValueX}{gemeenteValueY}")
```

```python
#same analysis for SoortRegio_2;
tempX = 'SoortRegio_2_x'
tempY = 'SoortRegio_2_y'
print(gemeenteValueX, gemeenteValueY)

for index, row in newMergeBenDis.iterrows():
    value_x = row[tempX]
    value_y = row[tempY]

    if value_x == value_y:
        print("looks good")
    else:
        print(f"look out for: {index} with {value_x} and {value_y} in {tempX}{tempY}")

#post analysis: values are the exact same in both columns; probably has to do with spacing
#
#most important to have correct is Codering_3_x and Codering_3_y. These are the neighourhood code values. By knowing from the output that
#they are the same in, means that the data per neighbourhood can be compared.
```

```python
#IndelingswijzigingWijkenEnBuurten_4_x and IndelingswijzigingWijkenEnBuurten_4_y analysis
tempX = 'IndelingswijzigingWijkenEnBuurten_4_x'
tempY = 'IndelingswijzigingWijkenEnBuurten_4_y'

for index, row in newMergeBenDis.iterrows():
    value_x = row[tempX]
    value_y = row[tempY]

    if value_x == value_y:
        print("looks good")
    else:
        print(f"look out for: {index} with {value_x} and {value_y} in {tempX}{tempY}")
#Post analysis: same values, except one input is float and the other an integer with a dot.
#They all have the same value: 1. Both columns will be dropped in the future, since they don't tell anything at all.
```
